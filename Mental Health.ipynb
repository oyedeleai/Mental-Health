```python
plt.figure(figsize=(15, 6))

# Create scatter plot
plt.scatter(df['Eating disorders (share of population) - Sex: Both - Age: Age-standardized'], 
            df['Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized'], 
            color='blue')

# Set labels
plt.xlabel("Eating disorders (share of population)")
plt.ylabel("Anxiety Disorders (share of population)")

# Set title
plt.title("Scatter Plot: Bipolar vs Eating Disorders")

# Show plot
plt.show()
```

```python

# Calculate the average share of population for Bipolar and Anxiety disorders
avg_bipolar = df['Bipolar disorders (share of population) - Sex: Both - Age: Age-standardized'].mean()
avg_eating = df['Eating disorders (share of population) - Sex: Both - Age: Age-standardized'].mean()

# Data for the bar plot
disorders = ['Bipolar Disorders', 'Eating Disorders']
avg_values = [avg_bipolar, avg_anxiety]

# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(disorders, avg_values, color=['blue', 'orange'])

# Set labels and title
plt.xlabel("Disorders")
plt.ylabel("Average Share of Population")
plt.title("Average Share of Population: Bipolar vs Eating Disorders")

# Show the plot
plt.show()
```

```python
avg_eating = df['Eating disorders (share of population) - Sex: Both - Age: Age-standardized'].mean()
avg_anxiety = df['Anxiety disorders (share of population) - Sex: Both - Age: Age-standardized'].mean()

# Data for the bar plot
disorders = ['Eating Disorders', 'Anxiety Disorders']
avg_values = [avg_eating, avg_anxiety]

# Plot the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(disorders, avg_values, color=['red', 'blue'])

# Set labels and title
plt.xlabel("Disorders")
plt.ylabel("Average Share of Population")
plt.title("Average Share of Population: Eating vs Anxiety Disorders")

# Add value labels on top of each bar
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, round(yval, 2), ha='center', va='bottom')

# Show plot
plt.show()
```

```python

# Calculate the average share of population for Bipolar and Anxiety disorders
avg_depressive = df['Depressive disorders (share of population) - Sex: Both - Age: Age-standardized'].mean()
avg_eating = df['Eating disorders (share of population) - Sex: Both - Age: Age-standardized'].mean()

# Data for the bar plot
disorders = ['Depressive Disorders', 'Eating Disorders']
avg_values = [avg_eating, avg_depressive]

# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(disorders, avg_values, color=['blue', 'orange'])

# Set labels and title
plt.xlabel("Disorders")
plt.ylabel("Average Share of Population")
plt.title("Average Share of Population: Depressive vs Eating Disorders")

# Show the plot
plt.show()
```

```python
# Calculate the average share of population for Bipolar and Anxiety disorders
avg_schizophrenia = df['Schizophrenia disorders (share of population) - Sex: Both - Age: Age-standardized'].mean()
avg_eating = df['Eating disorders (share of population) - Sex: Both - Age: Age-standardized'].mean()

# Data for the bar plot
disorders = ['Schizophrenia Disorders', 'Eating Disorders']
avg_values = [avg_eating, avg_schizophrenia]

# Plot the bar chart
plt.figure(figsize=(10, 6))
plt.bar(disorders, avg_values, color=['pink', 'purple'])

# Set labels and title
plt.xlabel("Disorders")
plt.ylabel("Average Share of Population")
plt.title("Average Share of Population: Schizophrenia vs Eating Disorders")

# Show the plot
plt.show()
```
